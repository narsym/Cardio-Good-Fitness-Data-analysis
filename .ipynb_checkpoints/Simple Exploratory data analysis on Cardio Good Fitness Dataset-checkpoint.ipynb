```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('CardioGoodFitness.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.pairplot(data)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(data['MaritalStatus'])
y = le.transform(data['MaritalStatus'])
```

```python
le1 = LabelEncoder()
le1.fit(data['Product'])
x = le1.transform(data['Product'])
```

```python
le2 = LabelEncoder()
le2.fit(data['Gender'])
z = le2.transform(data['Gender'])
gender = pd.DataFrame(x,range(len(z)),['Gender'])
gender.head()
```

```python
product = pd.DataFrame(x,range(len(x)),['Product'])
marital = pd.DataFrame(y,range(len(y)),['MaritalStatus'])
```

```python
tdata = data.drop(['Product','MaritalStatus'],axis = 1)
```

```python
tdata.head()
```

```python
fdata = pd.concat([tdata,product,marital],axis = 1)
```

```python
fdata.head()
```

```python
sns.pairplot(fdata)
```

```python
fcorr = fdata.corr()
sns.heatmap(fcorr,cmap = 'coolwarm',annot = True)
```

```python
sns.countplot(fdata['Product'])
```

```python
sns.swarmplot(x = 'Product',y = 'Fitness',data = fdata)
```

```python
sns.distplot(fdata['Income'])
```

```python
sns.boxplot(x = 'Gender',y = 'Product',data = fdata)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(pd.concat([fdata.drop(['Product','Gender'],axis = 1),gender],axis = 1),fdata['Product'])
```

```python
from sklearn import tree
print(tree.plot_tree(dtree))
```
